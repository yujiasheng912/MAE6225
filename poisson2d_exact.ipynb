```python
from math import pi
import numpy
%matplotlib inline
from matplotlib import pyplot
from matplotlib import rcParams
rcParams['font.family']='serif'
rcParams['font.size']=16
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import numba
from numba import jit
```

```python
def plot_3D(x,y,u):    
    fig = pyplot.figure(figsize=(11,7), dpi=100)
    ax = fig.gca(projection='3d')
    X,Y = numpy.meshgrid(x,y)
    surf = ax.plot_surface(X,Y,u[:], rstride=1, cstride=1, cmap=cm.viridis,
            linewidth=0, antialiased=False)

    ax.set_xlim(0,1)
    ax.set_ylim(0,1)
    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    ax.set_zlabel('$u$')
    ax.view_init(30,45)
```

```python
@jit
def poisson_sin_exact(ni,nj,n):
    dx=1./(ni-1)
    dy=1./(nj-1)
    x=numpy.linspace(0-dx/2,1+dx/2,ni+1)
    y=numpy.linspace(0-dy/2,1+dy/2,nj+1)
    X,Y=numpy.meshgrid(x,y)
    u_ex=numpy.sin(2*pi*n*X)*numpy.sin(2*pi*n*Y)
    plot_3D(x,y,u_ex)
    return u_ex
```

```python
@jit
def poisson_cos_exact(ni,nj,n):
    dx=1./(ni-1)
    dy=1./(nj-1)
    x=numpy.linspace(0-dx/2,1+dx/2,ni+1)
    y=numpy.linspace(0-dy/2,1+dy/2,nj+1)
    X,Y=numpy.meshgrid(x,y)
    u_ex=numpy.cos(2*pi*n*X)*numpy.cos(2*pi*n*Y)
    plot_3D(x,y,u_ex)
    return u_ex
```
