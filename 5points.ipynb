```python
import numpy
import sympy
sympy.init_printing()
phi,h1,h2,h3,h4,a,b,c,d,e=sympy.symbols('phi h_1 h_2 h_3 h_4 a b c d e')
```

```python
def f(h1,h2,h3,h4,n):
    return a*(h1+h2+h3+h4)**n+b*(h1+h2+h3)**n+c*(h1+h2)**n+d*h1**n
```

```python
sympy.solve([a+b+c+d+e,f(h1,h2,h3,h4,1)-1,f(h1,h2,h3,h4,2),f(h1,h2,h3,h4,3),f(h1,h2,h3,h4,4)],[a,b,c,d,e])
```
